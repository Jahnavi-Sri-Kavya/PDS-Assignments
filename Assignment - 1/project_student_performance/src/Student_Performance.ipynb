Score Distribution:

Are the scores normally distributed?

```python
# Plot histograms for math, reading, and writing scores
plt.figure(figsize=(15, 5))

# Math score distribution
plt.subplot(1, 3, 1)
sns.histplot(df['math score'], bins=30, kde=True, color='skyblue')
plt.title('Math Score Distribution')

# Reading score distribution
plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], bins=30, kde=True, color='salmon')
plt.title('Reading Score Distribution')

# Writing score distribution
plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], bins=30, kde=True, color='green')
plt.title('Writing Score Distribution')

plt.tight_layout()
plt.show()
```

Comparison of Average Scores between Genders:

```python
# Create a box plot for math scores by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', data=df, palette='pastel')
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()

# Create a box plot for reading scores by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='reading score', data=df, palette='pastel')
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()

# Create a box plot for writing scores by gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='writing score', data=df, palette='pastel')
plt.title('Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Writing Score')
plt.show()
```

Observations on Gender-Based Performance:

1. Math Scores:
   - Male students tend to score higher in math compared to female students.
   - This is evident from the higher median and quartiles in the box plot for math scores among male students.

2. Reading and Writing Scores:
   - Female students tend to score more in reading and writing compared to male students.
   - This observation is supported by higher median and quartiles in the box plots for reading and writing scores among female students.

These findings suggest a gender-based difference in performance across subjects, with male students excelling in math and female students performing better in reading and writing.


Relationship between Parental Education Levels and Student Scores:

```python
plt.figure(figsize=(15, 5))

# Math score by parental education level
plt.subplot(1, 3, 1)
sns.boxplot(x='parental level of education', y='math score', data=df, palette='pastel')
plt.title('Math Scores by Parental Education')
plt.xlabel('Parental Education Level')
plt.ylabel('Math Score')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Reading score by parental education level
plt.subplot(1, 3, 2)
sns.boxplot(x='parental level of education', y='reading score', data=df, palette='pastel')
plt.title('Reading Scores by Parental Education')
plt.xlabel('Parental Education Level')
plt.ylabel('Reading Score')
plt.xticks(rotation=45, ha='right')

# Writing score by parental education level
plt.subplot(1, 3, 3)
sns.boxplot(x='parental level of education', y='writing score', data=df, palette='pastel')
plt.title('Writing Scores by Parental Education')
plt.xlabel('Parental Education Level')
plt.ylabel('Writing Score')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()

# Show the plots
plt.show()
```

Observations on the Relationship Between Parental Education Levels and Student Scores:

1. Parents with a master's degree:
  - Their children tend to have the highest scores (70 and above) in math, reading, and writing.
  - This observation is consistent across all three subjects.

2. Parents with a high school education:
  - Their children tend to have the least scores across math, reading, and writing.
  - The scores are notably lower compared to other parental education levels.

These findings suggest a positive correlation between the level of parental education and the academic performance of students. Higher parental education levels, especially having a master's degree, appear to be associated with higher student scores, while lower education levels, such as having only a high school education, are associated with lower scores.


Impact of Test Preparation Course on Scores:

```python
# Create box plots for math, reading, and writing scores based on test preparation completion
plt.figure(figsize=(15, 5))

# Math score by test preparation completion
plt.subplot(1, 3, 1)
sns.boxplot(x='test preparation course', y='math score', data=df, palette='pastel')
plt.title('Math Scores by Test Preparation Completion')
plt.xlabel('Test Preparation Completion')
plt.ylabel('Math Score')

# Reading score by test preparation completion
plt.subplot(1, 3, 2)
sns.boxplot(x='test preparation course', y='reading score', data=df, palette='pastel')
plt.title('Reading Scores by Test Preparation Completion')
plt.xlabel('Test Preparation Completion')
plt.ylabel('Reading Score')

# Writing score by test preparation completion
plt.subplot(1, 3, 3)
sns.boxplot(x='test preparation course', y='writing score', data=df, palette='pastel')
plt.title('Writing Scores by Test Preparation Completion')
plt.xlabel('Test Preparation Completion')
plt.ylabel('Writing Score')

plt.tight_layout()

# Show the plots
plt.show()
```

Observations on the Impact of Test Preparation Course on Scores:

1. Math Scores:
  - Students who completed the test preparation course generally scored higher in math compared to those who did not.
  - This is evident from the higher median and quartiles in the box plot for math scores among students who completed the course.

2. Reading Scores:
  - Similar to math, students who completed the test preparation course tend to have higher reading scores.
  - The box plot for reading scores shows higher median and quartiles for students who completed the course.

3. Writing Scores:
  - The positive impact of the test preparation course is also observed in writing scores.
  - Students who completed the course have higher median and quartiles in the box plot for writing scores.

Overall, these findings suggest a noticeable improvement in scores for students who completed the test preparation course across all subjects. This implies that the course may be effective in enhancing student performance.


Impact of Test Preparation Course on Math Scores by Gender:

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='math score', hue='test preparation course', data=df, ci=None, palette='pastel')
plt.title('Impact of Test Preparation Course on Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.legend(title='Test Preparation Course')
plt.show()
```

Observations on the Impact of Test Preparation Course on Math Scores by Gender:
  
  - Male students who completed the test preparation course tend to have higher average math scores compared to those who did not.
  
  - Female students also show a positive trend, with higher average math scores for those who completed the test preparation course.

These observations suggest that the test preparation course may contribute to improved math performance for both male and female students.
